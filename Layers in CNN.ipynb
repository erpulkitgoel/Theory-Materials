Fully connected layer is used to classify data into multiple outputs. There we convert the 2D data into a 1D array and run the activation using horixzontal and vertical weights. The output of both the layers decide the class of the input data.

# D. Zero Padding


We add a border of Zero size pixels around the input image. So as not to lose data around the border.


- Valid :  It means there will be no padding and input size will not be maintained by padding during convolution
- SAME  :  It means the output will be of same size as the input size  as shown in above example. without losing valuabel information in the edges  during convolution.
