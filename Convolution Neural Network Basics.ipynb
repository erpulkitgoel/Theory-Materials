# Convolutional Neural Networks

CNN network ha two or more convolution layers in the network which are very efficient in detecting the patterns in the input.
A Convolution layer has a filter which  is an identity matrix of 3x3 or 5x5 etc which revolves or CONVOLVES around the input matrix to detect the patterns
Filter matrix is initialized with a random values as weights.

Filter generates the dot product of the 3x3 filter matrix with the block of pixels in the input image.

After calculating the dot product and sum of all products, the output is stored at the first place in output. This is how the output is having a smaller size as compared to the input which depends on the step size or STRIDE size.
