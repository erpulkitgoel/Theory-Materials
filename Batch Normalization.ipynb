# Batch normalization


There are two issues for why we use Normalization techniques.
- We have different ranges of data for different features. This causes the Exploding gradient descend problem
- Processing big values of input data in varing ranges causes more computational power.


Parameters "g" and "b" are updated during each epoch in each training batch.

So Batch Normalization is applied during the input stage and each convoltional layer outputs.
